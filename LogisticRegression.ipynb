```python

import numpy as np
import csv

maxy=0;

def costFunction(y,hyp):
    h1 = np.multiply(-1*y,np.log(hyp))
    h2 = np.multiply(1+(-1*y),np.log(1+(-1*hyp)))
    h3 = h1 - h2
    return np.sum(h3)
    

def sigmoid(z):
   
    return 1/(1+np.exp(-z))

def LogisticRegressionXfetcher():
    mylist = []
    with open("ex2data1.txt") as csvfile:
        reader = csv.reader(csvfile)
        mylist = list(reader)
    
    data = np.array(mylist,dtype='float64')
    x1,x2,y = data.T
    
    x1 = x1.reshape(100)
    x2 = x2.reshape(100)
    onesvector = np.ones(100)
    X = np.empty([3,100])
    X[0] = onesvector
    X[1] = x1
    X[2] = x2
   
    return X,y


       
theta = np.empty ([1,3])
X,y = LogisticRegressionXfetcher()
y = y.reshape(100)
m = X.shape[1]
i = 0
while i < 100000:
    
    z = np.dot(theta,X)
    hyp = sigmoid(z)
    hyp = hyp.reshape(100)
    
#     cost = np.sum(np.subtract(np.multiply(-1*y,np.log(hyp)),np.multiply(1-y,np.log(1-hyp))))/m
    cost = costFunction(y,hyp)/m
    print cost
    diff = np.subtract(hyp,y)
    diff = diff.reshape(1,100)
    core = np.dot(diff,X.T)  
    core = core * 0.005/m
    theta = theta - core 
    i+=1
    
print theta
```
